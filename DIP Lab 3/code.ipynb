```python
import cv2
import matplotlib.pyplot as plt

image_path = 'imgs/images.jpg'
image = cv2.imread(image_path)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)


plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')
plt.title("Binarized Image (Threshold = 127)")

plt.show()
```

# Task 3


```python
import cv2
import numpy as np

def manual_binarize(image, threshold):
    height, width = image.shape
    binarized_image = np.zeros((height, width), dtype=np.uint8)
    
    for i in range(height):
        for j in range(width):
            if image[i, j] > threshold:
                binarized_image[i, j] = 255
            else:
                binarized_image[i, j] = 0
    
    return binarized_image

input_image = cv2.imread('imgs/images.jpg')
height, width, channels = input_image.shape

if channels == 3:
    print("The image is a three-channel (color) image.")
    gray_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
    threshold_value = 128
    binarized_image = manual_binarize(gray_image, threshold_value)
    cv2.imshow('Binarized Image', binarized_image)
    
elif channels == 1:
    print("The image is a one-channel (grayscale) image.")
    cv2.imshow('Grayscale Image', input_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Task 4


```python
import cv2
import numpy as np

def manual_binarize(image, threshold):
    height, width = image.shape
    binarized_image = np.zeros((height, width), dtype=np.uint8)
    
    for i in range(height):
        for j in range(width):
            if image[i, j] > threshold:
                binarized_image[i, j] = 255
            else:
                binarized_image[i, j] = 0
    
    return binarized_image

input_image = cv2.imread('imgs/images.jpg')

gray_image = np.zeros((input_image.shape[0], input_image.shape[1]), dtype=np.uint8)

for i in range(input_image.shape[0]):
    for j in range(input_image.shape[1]):
        b, g, r = input_image[i, j]
        gray_intensity = int(0.299 * r + 0.587 * g + 0.114 * b)
        gray_image[i, j] = gray_intensity

threshold_value = 128
binarized_image = manual_binarize(gray_image, threshold_value)

cv2.imshow('Binarized Image', binarized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Task 5


```python
import cv2 as cv
 
cap = cv.VideoCapture(0)

while True:
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    cv.imshow('frame', gray)
    if cv.waitKey(1) == ord('q'):
        break
 
cap.release()
cv.destroyAllWindows()
```
