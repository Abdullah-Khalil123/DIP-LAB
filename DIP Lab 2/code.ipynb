```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

df['species'] = iris.target + 1

print("Iris Dataset Loaded:\n")
print(df.head(10)) 

print("\nDataset Information:")
print(df.info())

print("\nCount of Each Flower Type:")
print(df['species'].value_counts())

print("\nFull Iris Dataset:")
print(df)
```

```python
import cv2
import numpy as np

image = cv2.imread('image_task/data/image.jpg') 

blue_channel, green_channel, red_channel = cv2.split(image)

grayscale_image = (0.299 * red_channel + 0.587 * green_channel + 0.114 * blue_channel).astype(np.uint8)


cv2.imshow('Grayscale Image', grayscale_image)

plt.imshow(cv2.cvtColor(grayscale_image,cv2.COLOR_BGR2RGB))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
